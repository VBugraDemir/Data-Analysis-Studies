## Introduction to APIs

Web application programming interfaces, API's, the most common source of JSON data. It is a defined way for an application to communicate with other programs. They let programmers get data from an application without knowing database archirecture. But API's are shared resources and often limit how much data you can get in a specified timeframe.

Requests lets users send and get data from any url. So not tied to a particular API. Result is the response object, containing data and metadata.

response.json() returns a dictionary to convert it to a dataframe pd.DataFrame can be used.

headers keyword of requests takes a dictionary for user authentication to API.

## Working with nested JSONs

The data would be easier to work with if the nested attributes were in their own columns. pandas.io.json can do that. json_normalize returns a dataframe.

```python
from pandas.io.json import json_normalize
data = response.json()
cafes = json_normalize(data["businesses"], sep="_")
cafes.head()
```

The categories column is still nested.

### Handle deeply nested data


```python
flat_cafes = json_normalize(data["businesses"], sep="_", # coordinates are unnested now.
                           record_path="categories",
                           meta = ["name", "alias", "rating",
                                   ["coordinates", "latitude"],
                                   ["coordinates", "longitude"]],
                           meta_prefix="biz_")
flat_cafes.head()
```

## Combining multiple datasets

Project commonly draw data from multiple places.

Append adds rows from one data frame to another.

API commonly limit the number of records returned in a single call to manage resource usage.

Append and merge...

### Append data frames

APIs often limit the amount of data returned, since sending large datasets can be time- and resource-intensive. The Yelp Business Search API limits the results returned in a call to 50 records. However, the offset parameter lets a user retrieve results starting after a specified number.

```python
params = {"term":"cafe", "location":"NYC","sort_by":"rating","limit":50}
result = requests.get(api_url, headers=headers, params=params)
data = result.json()
top_50_cafes = json_normalize(data["businesses"])

params = {"term":"cafe", "location":"NYC","sort_by":"rating","limit":50,"offset":50}
result = requests.get(api_url, headers=headers, params=params)
data = result.json()
next_50_cafes = json_normalize(data["businesses"])

cafes = top_50_cafes.append(next_50_cafes, ignore_index=True) # it is important to reset index.
print(cafes.shape)
```

### Merge data frames

```python
crosswalk = pd.read_csv("crosswalk.csv", index_col=0)
pop_data = pd.read_csv("pop_data.csv", index_col=0, dtype={"puma":str})

cafes_with_pumas = cafes.merge(crosswalk, left_on="location.zip_code", 
                               right_on="zipcode")

cafes_with_pumas["puma"]=cafes_with_pumas["puma"].astype("int").astype("str")

cafes_with_pop = cafes_with_pumas.merge(pop_data ,on="puma")

cafes_with_pop.head()
```
