# Plotting Data with matplotlib

## Creating line plots

### Working hard

```python
import pandas as pd
import matplotlib.pyplot as plt
deshaun = pd.read_csv("deshaun.csv", index_col=0)
plt.plot(deshaun.day_of_week, deshaun.hours_worked)
plt.show()
```

## Adding text to plots

### Adding a legend


```python
plt.plot(deshaun.day_of_week, deshaun.hours_worked, label="Deshaun")
plt.plot(aditya.day_of_week, aditya.hours_worked, label="Aditya")
plt.plot(mengfei.day_of_week, mengfei.hours_worked, label="Mengfei")
plt.legend(loc="upper right")

plt.show()
```

### Adding labels

```python
plt.plot(deshaun.day_of_week, deshaun.hours_worked, label='Deshaun')
plt.plot(aditya.day_of_week, aditya.hours_worked, label='Aditya')
plt.plot(mengfei.day_of_week, mengfei.hours_worked, label='Mengfei')

plt.title("Hours worked per day, by officer, on the Missing Puppy Report for Bayes")
plt.ylabel("Hours worked per Day")
plt.legend(loc="upper right")
plt.show()
```

### Adding floating text

```python
six_months = pd.read_csv("six_months", index_col=0)

plt.plot(six_months.month, six_months.hours_worked)
plt.text(2.5, 80, "Missing June data")
plt.show()
```

## Styling graphs

Allowed color names are "web colors"

### Tracking crime statistics

```python
data = pd.read_csv("data.csv", index_col=0)
data["Year"] = data["Year"].astype("int")

plt.plot(data["Year"], data["Phoenix_Police_Dept"], label="Phoenix", color="DarkCyan")
plt.plot(data["Year"], data["Los_Angeles_Police_Dept"], label="Los Angeles", linestyle=":")
plt.plot(data["Year"], data["Philadelphia_Police_Dept"], label="Philadelphia", marker="s")
plt.legend()
plt.ylabel("Burglers per 100 Residents")
plt.xlabel("Year")
plt.xticks(data["Year"])
plt.show()
```

### Playing with styles

```python
plt.style.use("fivethirtyeight")
plt.plot(data["Year"], data["Phoenix_Police_Dept"], label="Phoenix", color="DarkCyan")
plt.plot(data["Year"], data["Los_Angeles_Police_Dept"], label="Los Angeles", linestyle=":")
plt.plot(data["Year"], data["Philadelphia_Police_Dept"], label="Philadelphia", marker="s")
plt.legend()
plt.ylabel("Burglers per 100 Residents")
plt.xlabel("Year")
plt.xticks(data["Year"])
plt.show()

plt.style.use("ggplot")
plt.plot(data["Year"], data["Phoenix_Police_Dept"], label="Phoenix", color="DarkCyan")
plt.plot(data["Year"], data["Los_Angeles_Police_Dept"], label="Los Angeles", linestyle=":")
plt.plot(data["Year"], data["Philadelphia_Police_Dept"], label="Philadelphia", marker="s")
plt.legend()
plt.ylabel("Burglers per 100 Residents")
plt.xlabel("Year")
plt.xticks(data["Year"])
plt.show()

plt.style.use("seaborn-notebook")
plt.plot(data["Year"], data["Phoenix_Police_Dept"], label="Phoenix", color="DarkCyan")
plt.plot(data["Year"], data["Los_Angeles_Police_Dept"], label="Los Angeles", linestyle=":")
plt.plot(data["Year"], data["Philadelphia_Police_Dept"], label="Philadelphia", marker="s")
plt.legend()
plt.ylabel("Burglers per 100 Residents")
plt.xlabel("Year")
plt.xticks(data["Year"])
plt.show()
print(plt.style.available)
```

### Identifying Bayes' kidnapper


```python
plt.style.use("ggplot")
suspect1 = pd.read_csv("suspect1.csv", index_col=0)
suspect2 = pd.read_csv("suspect2.csv", index_col=0)
ransom = pd.read_csv("ransom.csv", index_col=0)
plt.plot(ransom.letter, ransom.frequency, label="Ransom", linestyle=":", color="gray")
plt.plot(suspect1.letter, suspect1.frequency, label="Fred Frequentist")
plt.plot(suspect2.letter, suspect2.frequency, label="Gertrude Cox")
plt.xlabel("Letter")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

Fred Frequentist is the kidnapper. Both the ransom and Fred have low frequencies of H and high frequency of P.
