# Different Types of Plots

## Making a scatter plot

### Charting cellphone data

```python
import matplotlib.pyplot as plt
img = plt.imread("map.png")
fig, ax = plt.subplots()
img.resize((600,600))

ax.imshow(img)
```

```python
from PIL import Image
import numpy as np
import pandas as pd

cellphone = pd.read_csv("cellphone.csv",index_col=0)
cellphone.drop(["0", "Unnamed:"],axis=1)

img = Image.open("map.png")
img = img.resize((130,130))
a = np.asarray(img)
plt.imshow(a[::-1], origin='lower')

plt.scatter(cellphone.x+10, cellphone.y-5, color='red', marker='s',s=5, alpha=0.1)
plt.ylabel('Latitude')
plt.xlabel('Longitude')
plt.show()
```

## Making a bar chart

The best way to visualize a comparison of categorical data is usning a bar chart.

### Build a simple bar chart

```python
hours = pd.read_csv("hours.csv", index_col=0)
plt.bar(hours.officer, hours.avg_hours_worked, yerr=hours.std_hours_worked)
plt.show()
```

### Where did the time go?


```python
plt.bar(hours.officer, hours.desk_work, label="Desk Work")
plt.bar(hours.officer, hours.field_work, label="Field Work", bottom=hours.desk_work)
plt.legend()
plt.show()
```

## Making a histogram

By normalization two histograms can be made comparable even if the sample sizes are different.

### Modifying histograms

```python
puppies = pd.read_csv("puppies.csv", index_col=0)
plt.hist(puppies.weight)
plt.xlabel('Puppy Weight (lbs)')
plt.ylabel('Number of Puppies')
plt.show()

plt.hist(puppies.weight, bins=50)
plt.xlabel('Puppy Weight (lbs)')
plt.ylabel('Number of Puppies')
plt.show()

plt.hist(puppies.weight, range=(5,35))
plt.xlabel('Puppy Weight (lbs)')
plt.ylabel('Number of Puppies')
plt.show()
```

### Heroes with histograms


Distribution of gravel radii

```python
gravel = pd.read_csv("gravel.csv", index_col=0)
gravel = gravel.reset_index(drop=True).drop("0", axis=1)

plt.hist(gravel.radius)
plt.show()

plt.hist(gravel.radius,bins=40, range=(2,8))
plt.show()

plt.hist(gravel.radius,bins=40, range=(2,8), density=True)
plt.xlabel('Gravel Radius (mm)')
plt.ylabel('Frequency')
plt.title('Sample from Shoeprint')
plt.show()

gravel.radius.plot.kde()
plt.xlabel('Gravel Radius (mm)')
plt.ylabel('Frequency')
plt.title('Sample from Shoeprint')
plt.show()
```
