# Decorrelating your data and dimension reduction

PCA is often used before supervised learning to improve model performance and generalization. It can also be useful for unsupervised learning. For example, you'll employ a variant of PCA will allow you to cluster Wikipedia articles by their content!

## Visualizing the PCA transformation

More efficient storage and computation. Finds patterns in data and uses these patterns to re-express it in compressed form. This makes subsequent computation much more efficient. Remove less-informative noise features. 
It is used in many real-world applications.

It learns the "principal components" of the data.These are the directions in which the samples vary the most.

First step: decorrelation
Second step: reducing dimension

```python
import pandas as pd
import matplotlib.pyplot as plt
wine = pd.read_csv("wine.csv")
samples = wine[["total_phenols", "od280"]]
label = wine["class_label"]

from sklearn.decomposition import PCA
model = PCA()
model.fit(samples)
transformed = model.transform(samples)
plt.scatter("total_phenols", "od280", data = samples, c=label)

plt.scatter(transformed[:,0], transformed[:,1], c=label)
plt.axhline(0, c="black")
plt.axvline(0, c="black")
plt.show()
print(model.components_)
```

There is one row for each transformed sample.

### Correlated data in nature

```python
from scipy.stats import pearsonr
grains = pd.read_csv("seeds.csv", header=None)
grains_values = grains.values
width = grains_values[:,4]
length = grains_values[:,3]
plt.scatter(width, length)
plt.axis("equal")
plt.show()
correlation, pvalue = pearsonr(width, length)
print(correlation)
```

### Decorrelating the grain measurements with PCA

```python
grains = grains[[4,3]]
model = PCA()
pca_features = model.fit_transform(grains)
xs = pca_features[:,0]
ys = pca_features[:,1]
plt.scatter(xs, ys)
plt.axis("equal")
plt.show()
correlation, pvalue = pearsonr(xs, ys)
print(correlation)
```

## Intrinsic dimension

Intrinsic dimension informs the dimension reduction, because it tells us how much a dataset can be compressed. It is number of features needed to approcimate the dataset.

The intrinsic dimension is the number of PCA features that have significant variance.

### The first principal component

The first principal component of the data is the direction in which the data varies the most.

```python
plt.scatter(grains.values[:,0], grains.values[:,1])
model = PCA()
model.fit(grains)
mean = model.mean_
first_pc = model.components_[0,:]
plt.arrow(mean[0], mean[1], first_pc[0], first_pc[1], color="red", width=0.01)
plt.axis("equal")
plt.show()
```

### Variance of the PCA features

```python
fish = pd.read_csv("fish.csv", header=None)
fish_names = fish[0]
fish = fish.drop(0, axis=1)
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
scaler = StandardScaler()
pca = PCA()
pipeline = make_pipeline(scaler, pca)
pipeline.fit(fish)
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_)
plt.xlabel("PCA feature")
plt.ylabel("variance")
plt.xticks(features)
plt.show()
```

It looks like PCA features 0 and 1 have significant variance.

## Dimension reduction with PCA

Dimension reduction represents the same data using less features.

PCA features with lower variance are assumed to be noise.

Sparse arrays often represented using a special type of array called a csr_matrix. PCA doesn't support csr_matrices. Need to use TruncatedSVD.

### Dimension reduction of the fish measurements

```python
scaler = StandardScaler()
scaled_samples = scaler.fit_transform(fish)

pca = PCA(n_components = 2)
pca.fit(scaled_samples)
pca_features = pca.transform(scaled_samples)
print(pca_features.shape)
fish_names=fish_names.replace(['Bream', 'Roach', 'Smelt', 'Pike'],[0,1,2,3])
xs = pca_features[:, 0]
ys = pca_features[:, 1]
plt.scatter(xs, ys, c=fish_names)
plt.show()
```

```python
from sklearn.manifold import TSNE
model = TSNE(learning_rate=350)
tsne_features = model.fit_transform(fish)
xs = tsne_features[:,0]
ys = tsne_features[:,1]
plt.scatter(xs, ys, c=fish_names)
plt.title("TSNE without scaling")
plt.show()

from sklearn.manifold import TSNE
model = TSNE(learning_rate=250)
tsne_features = model.fit_transform(scaled_samples)
xs = tsne_features[:,0]
ys = tsne_features[:,1]
plt.scatter(xs, ys, c=fish_names)
plt.title("TSNE after scaling")
plt.show()
print("Variace of the unscaled data:", fish.values.var())
print("Variace of the scaled data:", scaled_samples.var())
```

### Clustering Wikipedia part I

TruncatedSVD is able to perform PCA on sparse arrays

```python
from scipy.sparse import csr_matrix
from sklearn.decomposition import TruncatedSVD
df = pd.read_csv("wikipedia-vectors.csv", index_col=0)
articles = csr_matrix(df.transpose())
titles = list(df.columns)
svd = TruncatedSVD(n_components=50)
svd_features = svd.fit_transform(articles)
features = range(len(svd.components_))

plt.bar(features, svd.explained_variance_)
plt.xlabel("PCA feature")
plt.ylabel("variance")
plt.xticks(features)
plt.show()

# default of n_components for TruncatedSVD is 2
from sklearn.cluster import KMeans
svd = TruncatedSVD(n_components=50)
kmeans = KMeans(n_clusters=6)
pipeline = make_pipeline(svd, kmeans)
```

### Clustering Wikipedia part II

```python
pipeline.fit(articles)
labels = pipeline.predict(articles)
df = pd.DataFrame({'label': labels, 'article': titles})
print(df.sort_values("label"))
```

```python
model = KMeans(n_clusters=6)
labels = model.fit_predict(articles)
df = pd.DataFrame({'label': labels, 'article': titles})
print(df.sort_values("label"))
```
