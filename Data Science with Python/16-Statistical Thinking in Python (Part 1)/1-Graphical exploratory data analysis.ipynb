# Graphical exploratory data analysis

## Plotting a histogram

### Plotting a histogram of iris data

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
data = load_iris()
versicolor_petal_length = data.data[data.target == 1][:,2]
```

```python
sns.set()
plt.hist(versicolor_petal_length)
plt.show()
```

### Axis labels!

```python
sns.set()
_ = plt.hist(versicolor_petal_length)
_ = plt.ylabel("count")
_ = plt.xlabel("petal length (cm)")
plt.show()
```

### Adjusting the number of bins in a histogram

```python
import numpy as np
n_data = len(versicolor_petal_length)
n_bins = np.sqrt(n_data)
n_bins = int(n_bins)
_ = plt.hist(versicolor_petal_length, bins=n_bins)
_ = plt.ylabel("count")
_ = plt.xlabel("petal length (cm)")
plt.show()
```

## Plot all of your data: Bee swarm plots

The problem with plotting histograms is that we are not plotting all of the data. We sweep the data into bins and losing their actual values.

### Bee swarm plot

```python
import pandas as pd
names = ["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)"]
df = pd.DataFrame(data.data, columns=names)
data.target_names
names = data.target[:]
l = []
for i in names:
    l.append(data.target_names[i])
namesdf = pd.DataFrame(l, columns=["species"])
df = df.join(namesdf)
```

```python
sns.swarmplot(x="species", y="petal length (cm)", data=df)
plt.xlabel("species")
plt.ylabel("petal length (cm)")
plt.show()
```

Virginica petals tend to be the longest, and setosa petals tend to be the shortest of the three species.

## Plot all of your data: ECDFs

ECDF shows all the data and gives a complete picture of how the data distributed.

### Computing the ECDF

```python
def ecdf(data):
    n = len(data)
    x = np.sort(data)
    y = np.arange(1, n+1) / n
    return x, y
```

### Plotting the ECDF

```python
x_vers, y_vers = ecdf(versicolor_petal_length)
plt.plot(x_vers, y_vers, marker=".", linestyle="none")
plt.xlabel("Petal length")
plt.ylabel("ECDF")
plt.show()
```

### Comparison of ECDFs


```python
setosa_petal_length = df[df["species"] == "setosa"]["petal length (cm)"]
x_set, y_set = ecdf(setosa_petal_length)

virginica_petal_length = df[df["species"] == "virginica"]["petal length (cm)"]
x_virg, y_virg = ecdf(virginica_petal_length)

plt.plot(x_set, y_set, marker=".", linestyle="none")
plt.plot(x_vers, y_vers, marker=".", linestyle="none")
plt.plot(x_virg, y_virg, marker=".", linestyle="none")
plt.legend(["setosa", "versicolor", "virginica"], loc="lower right")
plt.xlabel("petal length (cm)")
plt.ylabel("ECDF")
plt.show()
```

The ECDFs expose clear differences among the species. Setosa is much shorter, also with less absolute variability in petal length than versicolor and virginica.
