# Distributions

## Probability mass functions

### Make a PMF

```python
import pandas as pd
from empiricaldist import Pmf
```

```python
gss = pd.read_hdf("gss.hdf5")
pmf_year = empiricaldist.Pmf.from_seq(gss["year"], normalize=False)
print(pmf_year)
print(pmf_year[2016])
```

### Plot a PMF

```python
import matplotlib.pyplot as plt
age = gss["age"]
pmf_age = Pmf.from_seq(age, normalize=True)
pmf_age.bar()
plt.xlabel("Age")
plt.ylabel("PMF")
plt.show()

pmf_age.plot()
plt.xlabel("Age")
plt.ylabel("PMF")
plt.show()
a=gss.age.value_counts().sort_index()
plt.bar(x=a.index, height=a.values)
plt.show()
```

## Cumulative distribution functions

```python
# PMF: Probability that you get exactly x
# CDF: Probability that you get a value <= x for a given value of x

# IQR(interquartile range): the distance from the 25th to the 75th percentile. It measures the spead of the distribution,
# so it is similar to std or variance. Because it based on percentiles it doenst get thrown off by extereme values or
# outliers like the varience does. So IQR is more robust than variance, it works well even there are outliers or extreme 
# values.
```

### Make a CDF

```python
from empiricaldist import Cdf
```

```python
age = gss["age"]
cdf_age = Cdf.from_seq(age)
print(cdf_age(30))
# so fraction of the respondents that are older than 30 is 75%
```

### Compute IQR

```python
income = gss["realinc"]
cdf_income = Cdf.from_seq(income)
percentile_75th = cdf_income.inverse(0.75)
percentile_25th = cdf_income.inverse(0.25)
iqr = percentile_75th - percentile_25th
print(iqr)
```

### Plot a CDF

```python
cdf_income.plot()
plt.xlabel("Income (1986 USD)")
plt.ylabel("CDF")
plt.xticks([50000, 100000,150000])
plt.show()
```

## Comparing distributions

```python
# In general CDF's are smoother than PMF's
# For exploratory analysis CDFs are recommended
# The give a clear view of the distribution without too much noice
# and good for compating distributions.
```

### Distribution of education

```python
# The fraction of respondents with 12 years of education or fewer.
cdf_educ = Cdf.from_seq(gss["educ"])
cdf_educ[12]
```

### Extract education levels

```python
educ = gss["educ"]
bach = (educ >=16)
assc = (educ >=14) & (educ < 16)
high = (educ <= 12)
print(high.mean())
```

### Plot income CDFs

```python
income = gss['realinc']
Cdf.from_seq(income[high]).plot(label="High school")
Cdf.from_seq(income[assc]).plot(label="Associate")
Cdf.from_seq(income[bach]).plot(label="Bachelor")
plt.xlabel("Income (1986 USD)")
plt.ylabel("CDF")
plt.xticks([50000, 100000,150000])
plt.legend()
plt.show()
```

## Modeling distributions

```python
# When the data have a large number of unique values PMFs don't work very well.
# PDF is a more sensitive way to look for differences or too sensitive
# CDF to explore data
# PMF if there are small number of unique values
# KDE if there are a lot of values
```

### Distribution of income

```python
# In many datasets, the distribution of income is 
# approximately lognormal
import numpy as np
from scipy.stats import norm
income = gss["realinc"]
log_income = np.log10(income)

mean = log_income.mean()
std = log_income.std()
print(mean, std)
dist = norm(mean, std)
```

### Comparing CDFs

To see whether the distribution of income is well modeled by a lognormal distribution,
compare the CDF of the logarithm of the data to a normal distribution with 
the same mean and standard deviation.

```python
xs = np.linspace(2.3,5.2)
ys = dist.cdf(xs)
plt.plot(xs, ys, color ="gray")
Cdf.from_seq(log_income).plot()
plt.xlabel("log10 of realinc")
plt.ylabel("CDF")
plt.show()
```

The lognormal model is a pretty good fit for the data, but clearly not a perfect match.
That's what real data is like; sometimes it doesn't fit the model.



### Comparing PDFs

CDFs is used to see if the distribution of income is lognormal. The same comparison can be done by using a PDF and KDE.

```python
import seaborn as sns
xs = np.linspace(2.3,5.2)
ys = dist.pdf(xs)
plt.plot(xs, ys, color ="gray")
sns.kdeplot(log_income)
plt.xlabel("log10 of realinc")
plt.ylabel("PDF")
plt.show()
```
